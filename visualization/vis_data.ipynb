```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import os
os.chdir('..')
%matplotlib notebook
%matplotlib inline
```

```python
input_all = []
output_all = []
input_train = []
output_train = []

with open("raw_data/values_FAB.txt", 'r') as v:
    lines = v.readlines()
for i in lines:
    i = list(map(float, eval(i)))
    
    input_all.append(i[:2])
    output_all.append(i[2:4])  
    
with open("data_splits/values_FAB_train.txt", 'r') as v:
    lines = v.readlines()
for i in lines:
    i = list(map(float, eval(i)))
    
    input_train.append(i[:2])
    output_train.append(i[2:4])  
    
```

```python
X_all = np.array([i[0] for i in input_all])

Y_all = np.array([i[1] for i in input_all])

Z_1_all = np.array([i[0] for i in output_all])
Z_2_all = np.array([i[1] for i in output_all])
###################################################

X_train = np.array([i[0] for i in input_train])

Y_train = np.array([i[1] for i in input_train])

Z_1_train = np.array([i[0] for i in output_train])
Z_2_train = np.array([i[1] for i in output_train])
```

```python

fig = plt.figure(figsize=(20,20))

ax = fig.add_subplot(121, projection='3d')
ax.scatter(X_all, Y_all, Z_1_all, marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


ax = fig.add_subplot(122, projection='3d')
ax.scatter(X_train, Y_train, Z_1_train, marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
